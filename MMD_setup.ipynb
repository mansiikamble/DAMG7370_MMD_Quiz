```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark.sql("USE CATALOG `workspace`")
spark.sql("USE SCHEMA `materialmasterdetail`")
```

```python
df = spark.read.option(
    "inferSchema",
    "true",
).option(
    "header",
    "true",
).option(
    "sep",
    "|",

).csv(
    "/Volumes/workspace/materialmasterdetail/datastore/material_master_1k.csv"
)

df.write.mode("overwrite").saveAsTable("workspace.materialmasterdetail.material_master")
```
